```python
df1=df.drop(['Index','Description','Status', 'Transaction', 'Furnishing', 'facing', 'overlooking', 'Society',
             'Balcony', 'Car Parking', 'Ownership', 'Super Area','Dimensions', 'Plot Area','Floor'],axis='columns')
df1.head()
```

```python
df1.head()
```

```python
df1.isnull().sum()
```

```python
df2 = df1.dropna()
df2.isnull().sum()
```

```python
df2.head()
```

```python
df2.shape
```

```python
import re

df2['Bhk'] = df2['Title'].str.extract(r'(\d+)\s*BHK') 
```

```python
df2.head()
```

```python
df3 = df2.drop(['Title'],axis=1)
df3.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import LabelEncoder
```

```python
#le = LabelEncoder()
#x['location'] = le.fit_transform(x['location'])
#x.head()
```

```python
df6 = df3['location']
df6.head()
```

```python
df6 = pd.get_dummies(df6, columns=['location'], drop_first=True)
df6 = df6.astype(int)
df6.head()
```

```python
df7 = pd.concat([df3,df6],axis='columns')
```

```python
df8 = df7.drop(['location','Price (in rupees)'],axis='columns')
df8.head()
```

```python
df8['Carpet Area'] = df8['Carpet Area'].str.extract('(\d+)').astype(int)
```

```python
df8['Amount(in rupees)'] = df8['Amount(in rupees)'].str.extract(r'(\d+\.?\d*)').astype(float)
```

```python
df8 = df8.dropna()
df8.head()
```

```python
df8
```

```python
x = df8.drop('Amount(in rupees)',axis = 'columns')
x.head()
import pickle
pickle.dump(x.columns.tolist(), open("model_columns.pkl", "wb"))
```

```python
y = df8['Amount(in rupees)']
y
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=10)
```

```python
df8['Bathroom'] = df8['Bathroom'].replace({'> 10': 11}).astype(int)
```

```python
df8['Bathroom'].unique()
```

```python
lr = LinearRegression()
lr.fit(x_train,y_train)
lr.score(x_test,y_test)
```

```python
yp = lr.predict(x_test)
yp
```

```python
from sklearn.metrics import confusion_matrix,f1_score,mean_absolute_error,r2_score,recall_score,roc_auc_score,roc_curve,mean_squared_error
```

```python
print(mean_absolute_error(y_test,yp))
print(mean_squared_error(y_test,yp))
```

```python
def predict_price(location, carpet_area, bathroom, bhk):
    loc_index = np.where(x.columns == location)[0][0]

    x1 = np.zeros(len(x.columns))
    x1[0] = carpet_area
    x1[1] = bathroom
    x1[2] = bhk
    if loc_index >= 0:
        x1[loc_index] = 1

    # Pass as DataFrame with correct column names
    input_df = pd.DataFrame([x1], columns=x.columns)

    # Predict and return as normal Python float rounded to 2 decimals
    predicted_price = rfr.predict(input_df)[0]
    return float(round(predicted_price, 2))
```
